```python
def electro(videocards, result, boarder):
    for i in range(len(videocards) ):
        arr = []

        for j in range(len(videocards) ):
            P = 0
            N = 0

            if i == j:
                arr.append(0)
                continue

            if int(videocards[i][1]) < int(videocards[j][1]):
                P += 5
            elif int(videocards[i][1]) > int(videocards[j][1]):
                N += 5

            if int(videocards[i][2]) < int(videocards[j][2]):
                P += 3
            elif int(videocards[i][2]) > int(videocards[j][2]):
                N += 3

            if int(videocards[i][3]) > int(videocards[j][3]):
                P += 4
            elif int(videocards[i][3]) < int(videocards[j][3]):
                N += 4

            if int(videocards[i][4]) > int(videocards[j][4]):
                P += 3
            elif int(videocards[i][4]) < int(videocards[j][4]):
                N += 3

            print("P", "<",i + 1, j + 1, ">", P)
            print("N", "<",i + 1, j + 1, ">", N)

            if P == 0 or (P == 0 and N == 0):
                arr.append(math.inf)
            elif N == 0 or P == N:
                arr.append(0)
            else:
                D = round(P / N, 1)
                print("D", "<",i + 1, j + 1, ">", D)
                if D <= boarder:
                    arr.append(0)
                else:
                    arr.append(D)

        result[videocards[i][0]] = arr

    return result
```

пометка:

Список "res" сортируется по первому элементу в каждом списке
(то есть по количеству ненулевых элементов) в порядке убывания с помощью параметра "key".

```python
def list_of_the_best(table):
    res = []
    matrix = np.array(table)

    for i in range(0, len(matrix)):
        arr_videocards_float = list(map(float, matrix[i]))
        arr_videocards_no_zero = [i for i in arr_videocards_float if i != 0.0]
        elem_res = [len(arr_videocards_no_zero), i + 1]
        res.append(elem_res)

    res = sorted(res, key=lambda weight: weight[0], reverse=True)
    print("res", res)

    print('Selection:')
    iteration = 0
    for finaly in res:
        iteration+=1
        print(iteration, ') Альтернатива', finaly[1])
```
